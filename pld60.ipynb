# pralados60 study

The subjects were 'randomly' assigned all at once before the training program to one of the two groups, Walk and Run, by the method of minimisation (see [Saguaei (2011)](http://www.ncbi.nlm.nih.gov/pmc/articles/PMC3317766/)) using the factors gender, BMI (below/over 28.3 kg/m$2$), and then $\dot{V}O_2$ peak (below/over 22.6 ml/kg/min). We also set the assignment with 30% more subjects for the Run group. This assignment was performed using the [Minim software](http://www-users.york.ac.uk/~mb55/guide/minim.htm).

We started with **15** subjects in the Walk group and **19** subjects in the Run group.   
At the end, we had **12** subjects in the Walk group and **10** subjects in the Run group.

Here is the statistical data analysis.

```python
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from IPython.display import HTML, display, Image
from factorplots2 import interaction_plot
```

### Comparisons for age, mass, height, and BMI

Let's test if these two groups of *survivals* are similar about age, mass, height, and BMI.   

**The data:**

```python
# mass before, mass after, height, age
r = np.array([[63.1, 62.2, 1.62, 64],
             [88.9, 84.7, 1.66, 64],
             [64.5, 64.5, 1.52, 71],
             [55.0, 54.3, 1.63, 71],
             [73.7, 73.9, 1.71, 62],
             [89.4, 91.1, 1.77, 66],
             [87.9, 88.1, 1.56, 61],
             [84.3, 83.5, 1.73, 58],
             [79.2, 79.2, 1.52, 66],
             [78.2, 78.2, 1.56, 60],
             [76.8, 76.8, 1.63, 64],
             [64.7, 64.7, 1.64, 60]])
# mass before, mass after, height, age
w = np.array([[74.5, 74.1, 1.64, 68],
             [87.9, 87.5, 1.57, 71],
             [71.9, 71.9, 1.62, np.nan],
             [68.5, 71.7, 1.68, 62],
             [105.4, 105.4, 1.73, 67],
             [71.7, 70.5, 1.47, 64],
             [60.7, 60.5, 1.52, 64],
             [83.1, 82.6, 1.65, 68],
             [74.6, 74.6, 1.62, 68],
             [61.4, 61.4, 1.49, 57]])
data = np.vstack((w, r))
# add BMI data
bmi = lambda weight, height: weight / height**2
data = np.vstack((data[:, :3].T, bmi(data[:, 0], data[:, 2]), 
                  bmi(data[:, 1], data[:, 2]), data[:, 3].T)).T
```

#### Descriptive statistics:

```python
columns = ['Mass0', 'Mass1', 'Height', 'BMI0', 'BMI1', 'Age']
col_order = np.argsort(columns)
df = pd.DataFrame(data, columns=columns)
df['Group'] = pd.Series(['Walk']*w.shape[0] + ['Run']*r.shape[0])
#display(df.groupby('Group').describe())
#axs = df.boxplot(by='Group', figsize=(10, 6))
#plt.tight_layout()

plt.figure(figsize=(10, 5))
plt.subplot(2, 3, 1)
interaction_plot(x=df.Group, trace=df.Group, response=df.Mass0, legendloc=None)
plt.xlabel(''), plt.tick_params(labelbottom=False)
plt.subplot(2, 3, 2)
interaction_plot(x=df.Group, trace=df.Group, response=df.Mass1, legendloc=None)
plt.xlabel(''), plt.tick_params(labelbottom=False)
plt.subplot(2, 3, 3)
interaction_plot(x=df.Group, trace=df.Group, response=df.Height, legendloc=None)
plt.xlabel(''), plt.tick_params(labelbottom=False)
plt.subplot(2, 3, 4)
interaction_plot(x=df.Group, trace=df.Group, response=df.BMI0, legendloc=None)
plt.subplot(2, 3, 5)
interaction_plot(x=df.Group, trace=df.Group, response=df.BMI1, legendloc=None)
plt.subplot(2, 3, 6)
interaction_plot(x=df.Group, trace=df.Group, response=df.Age, legendloc=None)
plt.suptitle('Age, mass, height, and BMI', y=1.04, size=20)
plt.tight_layout()
```

#### t-tests for the between-group comparisons:

```python
from scipy import stats
walk = df[df['Group'] == 'Walk']
run  = df[df['Group'] == 'Run']
print('Walk versus Run (t-test)')
for c in columns:
    t, p = stats.ttest_ind(walk[c].dropna(), run[c].dropna())
    print('Factor %s: t=%.2f, p=%.2f' %(c, t, p))
#stats.ttest_re
```

**In conclusion, there are no difference between groups with respect to age, mass, height, and BMI.**

```python
# load IPython extension to work with R
%load_ext rmagic
```

```python
%%R -i data
data$Subject = factor(data$Subject)
library(car)
library(ez)
options(contrasts=c("contr.sum", "contr.poly"))
print("Variable: Str_dur")
# test homogeneity of variance between groups
r = leveneTest(data$Str_dur, data$Group, center="mean")
print(r)
r = ezANOVA(data=data, dv=.(Str_dur), wid=.(Subject), within=.(Period), between=.(Group), type=3)
print(r)
print("Variable: Str_len")
# test homogeneity of variance between groups
r = leveneTest(data$Str_len, data$Group, center="mean")
print(r)
r = ezANOVA(data=data, dv=.(Str_len), wid=.(Subject), within=.(Period), between=.(Group), type=3)
print(r)
print("Variable: MIF")
# test homogeneity of variance between groups
r = leveneTest(data$MIF, data$Group, center="mean")
print(r)
r = ezANOVA(data=data, dv=.(MIF), wid=.(Subject), within=.(Period), between=.(Group), type=3)
print(r)
print("Variable: Pks1")
# test homogeneity of variance between groups
r = leveneTest(data$Pks1, data$Group, center="mean")
print(r)
r = ezANOVA(data=data, dv=.(Pks1), wid=.(Subject), within=.(Period), between=.(Group), type=3)
print(r)
print("Variable: Pks2")
# test homogeneity of variance between groups
r = leveneTest(data$Pks2, data$Group, center="mean")
print(r)
r = ezANOVA(data=data, dv=.(Pks2), wid=.(Subject), within=.(Period), between=.(Group), type=3)
print(r)
print("Variable: Pks3")
# test homogeneity of variance between groups
r = leveneTest(data$Pks3, data$Group, center="mean")
print(r)
r = ezANOVA(data=data, dv=.(Pks3), wid=.(Subject), within=.(Period), between=.(Group), type=3)
print(r)
```

**Post-hoc analysis**   
In our 2x2 design, we are interested in only four comparisons for the interaction effect: Pre:G1/G2; Pos:G1/G2; G1:Pre/Pos; G2:Pre/Pos and two comparisons for each of the main effects.   
So, as post hoc, let's use t-tests with Bonferroni correction (the p value will be multiplied by 4 for the interaction effects and by 2 for the main effects).

```python
%%R -i data
print(t.test(subset(data$Str_len, data$Group=="walk" & data$Period=="pre"),
             subset(data$Str_len, data$Group=="run"  & data$Period=="pre")))
print(t.test(subset(data$Str_len, data$Group=="walk" & data$Period=="pos"), 
             subset(data$Str_len, data$Group=="run"  & data$Period=="pos")))
print(t.test(subset(data$Str_len, data$Group=="walk" & data$Period=="pre"), 
             subset(data$Str_len, data$Group=="walk" & data$Period=="pos"), paired=T))
print(t.test(subset(data$Str_len, data$Group=="run"  & data$Period=="pre"), 
             subset(data$Str_len, data$Group=="run"  & data$Period=="pos"), paired=T))
print("effect size")
library(lsr)
A = data$Str_len[data$Group=="run"]
B = data$Period[data$Group=="run"]
print(cohensD(A ~ B), paired=T)
```

```python
%%R -i data
print(t.test(subset(data$Pks2, data$Group=="walk" & data$Period=="pre"),
             subset(data$Pks2, data$Group=="run"  & data$Period=="pre")))
print(t.test(subset(data$Pks2, data$Group=="walk" & data$Period=="pos"), 
             subset(data$Pks2, data$Group=="run"  & data$Period=="pos")))
print(t.test(subset(data$Pks2, data$Group=="walk" & data$Period=="pre"), 
             subset(data$Pks2, data$Group=="walk" & data$Period=="pos"), paired=T))
print(t.test(subset(data$Pks2, data$Group=="run"  & data$Period=="pre"), 
             subset(data$Pks2, data$Group=="run"  & data$Period=="pos"), paired=T))
```

**In conclusion, during walking there were statistically significant interactions between factors Group and Period for the variables stride length (p=0.046, $\mathbf{\eta2=0.01}$) and the first active peak of the vertical ground reaction force (p=0.009, $\mathbf{\eta2=0.02}$). The post-hoc analysis revealed that only the variable stride length for the group 'run' increased after training (p=0.047, d=0.33). See the corresponding interaction plots above. **
