According to Mayara:

- São três testes que fiz o registro da EMG. Esse arquivo é de um teste progressivo até a exaustão nos momentos pré (1) e pós (2) treinamento.
- A comparação que pensei em fazer é pré e pós do grupo treinamento e pré e pós do grupo controle. Acho que não tem sentido ficar comparando mais do que isso. Nos outros testes as comparações são as mesmas, o que vai mudar é o número de estágios.
- Nesse sinal do teste progressivo (aumento da velocidade a cada 3 minutos), o registro foi feito nos 10 segundos finais de cada estágio.
- Siglas: E1V12 - "E" é a sigla para o sujeito, "1" é o momento (pré treinamento) e "V12" é a velocidade do estágio. Esse sujeito foi até o estágio de 18km.h tanto no pré quanto no pós treinamento. Porém em alguns casos eu não vou ter a mesma quantidade de registros pré e pós (às vezes pela diferença na velocidade atingida e às vezes por perda do sinal).

```python
# Import the necessary libraries
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.signal import butter, filtfilt
from threshold_detector import threshold_detector
from linear_envelope import linear_envelope
```

```python
# file with 3 columns: time, EMG VMO, EMG BF
# coding for file names:
tasks = ['exhaustion', '1RM', 'limit_time']
subjects = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'O', 'P', 'Q']
instants = ['1', '2']
speeds = ['V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18']
extension = '.txt'
path2 = r'../X/Mayara/'  # file directory
```

```python
# load files and plot raw data
for subject in subjects:
    fig, ax = plt.subplots(7, 2, sharex=True, squeeze=True, figsize=(14, 10))
    for i, instant in enumerate(instants):
        # MVIC data
        """
        filename = path2 + tasks[1] + '/' + subject + instant + extension
        try:
            data = np.loadtxt(filename, skiprows=5, unpack=False)
            time, VMOm, BFm = data[:, 0], data[:, 1], data[:, 2]
            fs = np.mean(1/np.diff(time))
            VMOm = linear_envelope(VMOm, freq=fs, Fc_bp=[20, 400], Fc_lp=1)       
            BFm = linear_envelope(BFm, freq=fs, Fc_bp=[20, 400], Fc_lp=1)
            VMOm, BFm = np.max(VMOm), np.max(BFm)
        except:
            VMOm, BFm = 1, 1
            print('File ' + subject + instant + speed + extension + ' does not exist.')
        """    
        VMOm, BFm = 1, 1
                
        for s, speed in enumerate(speeds):
            filename = path2 + tasks[0] + '/' + subject + instant + speed + extension
            time = VMO = BF = FO = np.array([])
            try:
                data = np.loadtxt(filename, skiprows=5, unpack=False)               
                if data.shape[1] == 2:
                    time, BF = data[:, 0], data[:, 1]
                elif data.shape[1] == 3:
                    time, VMO, BF = data[:, 0], data[:, 1], data[:, 2]
                elif data.shape[1] == 4:
                    time, VMO, BF, FO = data[:, 0], data[:, 1], data[:, 2], data[:,3]                    
            except:
                #pass
                print('File ' + subject + instant + speed + extension + ' does not exist.')
            
            fs = np.mean(1/np.diff(time))
            b, a = butter(4, np.array([20, 400])/(fs/2), btype = 'bandpass')
            
            if len(VMO):
                VMO = filtfilt(b, a, VMO) / VMOm
                ax[s, i].plot(time, VMO, color = [1, 0, 0, .5], label='VMO')
            if len(BF):    
                BF = filtfilt(b, a, BF) / BFm
                ax[s, i].plot(time, BF, color = [0, 0, 1, .5], label='BF')
            if len(FO):
                inds = threshold_detector(FO, threshold=1)
                for (indi, indf) in inds:
                    ax[s, i].plot(np.array([time[indi], time[indi]]), ax[s, i].set_ylim(), 'g')
                    ax[s, i].plot(np.array([time[indf], time[indf]]), ax[s, i].set_ylim(), 'g')
            
            ax[s, i].locator_params(axis = 'y', tight=True, nbins = 4)
            ax[s, i].set_xlim((0, 10))
            ax[s, 0].set_ylabel(speed, fontsize=14)
            
    if len(BF):
        ax[0, 0].legend(loc=1, framealpha=.5, title='Muscles')
        ax[0, 1].legend(loc=1, framealpha=.5, title='Muscles')
    ax[0, 0].set_title('PRE')
    ax[0, 1].set_title('POST')
    ax[6, 0].set_xlabel('Time [s]', fontsize=14)
    ax[6, 1].set_xlabel('Time [s]', fontsize=14)
    fig.suptitle('Subject ' + subject + ': muscle activation', fontsize=18)
    fig.subplots_adjust(hspace=0)
    plt.show()
```
